## Using ANOVA to verify if this values have an impact on saleprice

```python
def NullHypoTest(data: pd.DataFrame, value: str, cols: list):
    for col in cols:
        model = ols(f"{value} ~ C(data[{col}])" , data = data).fit()
        anova_table = sm.stats.anova_lm(model, typ=2)
        p = anova_table['PR(>F)'][0]
        if p < 0.5:
            print(f"Column {col}:\n  p_value ({p}) > 0.05. Price is affected")
```
